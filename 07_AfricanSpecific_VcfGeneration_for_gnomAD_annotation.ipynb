```python
import os
import sys
import pandas as pd
```

```python
def create_analysis_folder(operating_system, project_name, analysis_name):
	
	# Set path to BSC folder for mac, linux and windows
	if operating_system == 'windows':
		pathToBSCfolder = 'Z:\\'
	elif operating_system == 'linux':
		pathToBSCfolder = os.path.join('/home','amontaner','BSC')
	elif operating_system == 'mac':
		pathToBSCfolder = os.path.join('/Users','annamontaner','Documents','BSC3')
	else:
		print ('undefined operating system') 
	
	# Path to project folder 
	project_folder = os.path.join('scratch','cli79','cli79334','projects', 'other', project_name)


   # Create analysis folder and subfolders 
	analysis_folder = os.path.join(pathToBSCfolder, project_folder, analysis_name)
	input_data = os.path.join(analysis_folder,'input_data')
	output_data = os.path.join(analysis_folder,'output_data')


	folder_list = [analysis_folder,
				   input_data, 
				   output_data]

	if not os.path.exists(analysis_folder):
		for items in folder_list:
			os.makedirs(items)
		print("Analysis folder and subfolders created '{}' created".format(analysis_name))
	else: 
		print("Already existing analysis '{}' ".format(analysis_name))

	return(analysis_folder)
```

```python
# Create analysis folders
analysis_folder = create_analysis_folder('mac','1kGPhg38_african_variants_abundant_specific', '01_vcf_generation')	
```

```python
# Copy imported dataframe
afr_specific = imported.copy()
```

```python
# Build columns required for vcf
afr_specific['QUAL'] = "."
afr_specific['FILTER'] = "NONE"
afr_specific['#CHROM'] = afr_specific['CHROM']
afr_specific['INFO'] = 'fedorova_label='+afr_specific['fedorova_label']+";adme_label="+afr_specific['ADME_label']+";abundantAFR_label="+afr_specific['abundantAFR_label']+";maxFreqs="+afr_specific['maxFreqs'].astype(str)+";AFR_overrepresentation="+afr_specific['AFR_overrepresentation'].astype(str)+";AFR_overrepresentation_mod="+afr_specific['AFR_overrepresentation_mod'].astype(str)+";AFR_overrepresentation_label="+afr_specific['AFR_overrepresentation_label'].astype(str)+";specificAFR_label="+afr_specific['specificAFR_label']
```

```python
# Generate .tsv to export
afr_specific_for_vcf = afr_specific[['#CHROM','POS','ID','REF','ALT','QUAL','FILTER','INFO']]
```

```python
# Export .tsv
afr_specific_for_vcf.to_csv(os.path.join(analysis_folder,'output_data','afr_specific_for_vcf.tsv'), index=False, sep="\t")
```

```python
#### Generate vcf for SnpEff annotation
input_TSV = os.path.join(os.path.join(analysis_folder,'output_data',"afr_specific_for_vcf.tsv"))
output_VCF = os.path.join(os.path.join(analysis_folder,'output_data',"afr_specific.vcf"))

I = open(input_TSV, "r")
O = open(output_VCF, "w")

O.write("##fileformat=VCFv4.0\n")
O.write("##FILTER=<ID=ID,Number=1,Type=String,Description=\"dbSnp id\">\n")
O.write("##INFO=<ID=fedorova_label,Number=1,Type=String,Description=\"Present or not present in Fedorova data\">\n")
O.write("##INFO=<ID=adme_label,Number=1,Type=String,Description=\"ADME or not ADME genes\">\n")
O.write("##INFO=<ID=abundantAFR_label,Number=1,Type=String,Description=\"Variant status: abundant or not abundant in African populations in the 1kGPhg38 dataset\">\n")
O.write("##INFO=<ID=maxFreqs,Number=1,Type=Float,Description=\"Maximum frequency observed in populations other than African in the 1kGPhg38 dataset\">\n")
O.write("##INFO=<ID=AFR_overrepresentation,Number=A,Type=Float,Description=\"African specificity raw\">\n")
O.write("##INFO=<ID=AFR_overrepresentation_mod,Number=A,Type=Float,Description=\"African specificity modified\">\n")
O.write("##INFO=<ID=AFR_overrepresentation_label,Number=1,Type=String,Description=\"African specificity as factor\">\n")
O.write("##INFO=<ID=specificAFR_label,Number=1,Type=String,Description=\"Variant status: specific or not specific to African populations in the 1kGPhg38 dataset\">\n")
O.write("##FORMAT=<ID=NO,Number=1,Type=String,Description=\"Not used\">\n")


for iLine in I:
	list = iLine.split("\t")
	newLine = '\t'.join(list)
	O.write(newLine)

I.close()
O.close()
```
